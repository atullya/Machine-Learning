Gradient descent is an algorithm that finds the best fit line for the given training data set

### PARTIAL DERIVATIVE OF 'm' and 'b' 


```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    iteration=1000
    n=len(x) #number of data points
    learning_rate=0.01 #learning rate
    for i in range(iteration):
        y_predict=m_curr*x+b_curr
        cost=(1/n)*sum([val**2 for val in (y-y_predict)])
        md=-(2/n)*sum(x*(y-y_predict))
        bd=-(2/n)*sum(y-y_predict)
        m_curr=m_curr-md*learning_rate 
        b_curr=b_curr-bd*learning_rate
        print("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost,i))
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)
```
