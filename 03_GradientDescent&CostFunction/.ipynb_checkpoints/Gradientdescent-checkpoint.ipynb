### PARTIAL DERIVATIVE OF 'm' and 'b' 





```python
import numpy as np
```

```python
import numpy as np


def gradient_descent(x,y):
    m_curr=b_curr=0
    iteration=1000
    learning_rate=0.01 #learning rate
    n=len(x)
    for i in range(iteration):
        y_predict=m_curr*x+b_curr
        cost=(1/n)*sum([val**2 for val in (y-y_predict)])
        md=-(2/n)*sum(x*(y-y_predict))
        bd=-(2/n)*sum(y-y_predict)
        m_curr=m_curr-md*learning_rate 
        b_curr=b_curr-bd*learning_rate
        print("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost,i))


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 5, 6])  # some random data points

# Line equation for demonstration: y = 1x + 2
y_line = 1 * x + 2

plt.figure(figsize=(6, 5))

# Scatter plot - data points
plt.scatter(x, y, color='blue', label='Actual Data Points', s=100)

# Line plot - model prediction or trend
plt.plot(x, y_line, color='red', label='Prediction Line (y = x + 2)', linewidth=2)

plt.title('Scatter vs Plot Example')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
