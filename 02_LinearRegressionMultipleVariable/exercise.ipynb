```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 
```

```python
df=pd.read_csv('hiring.csv')
df
```

```python
df.columns
```

```python
df.rename(columns={'test_score(out of 10)':'test_score','interview_score(out of 10)':'interview_score','salary($)':'salary'},inplace=True)
```

```python
df.test_score.median()
```

```python
df.test_score=df.test_score.fillna(df.test_score.median())
df
```

```python
import word2number
from word2number import w2n
```

```python
df.experience=df.experience.fillna('zero')
df
```

```python
from word2number import w2n

# Ensure all values in 'experience' are strings
df['experience'] = df['experience'].astype(str)

# Replace 'nan' or missing values with 'zero' before conversion
df['experience'] = df['experience'].replace('nan', 'zero')

# Convert words to numbers
df['experience'] = df['experience'].apply(w2n.word_to_num)

print(df)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['experience','test_score','interview_score']],df['salary'])
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```

```python
import ipywidgets as widgets
from IPython.display import display
import numpy as np

# Input widgets
area_input = widgets.FloatText(description='Area (sq ft):', value=2600)
calculate_button = widgets.Button(description="Predict Price")

# Output widget
output = widgets.Output()

# Linear regression values from previous calculation
m = 135.82
c = 180510.4

def predict_price(area):
    return m * area + c

# Define button click behavior
def on_calculate_clicked(b):
    with output:
        output.clear_output()
        area = area_input.value
        price = predict_price(area)
        print(f"Predicted Price: ${price:,.2f}")

calculate_button.on_click(on_calculate_clicked)

# Show the interface
display(area_input, calculate_button, output)
```
