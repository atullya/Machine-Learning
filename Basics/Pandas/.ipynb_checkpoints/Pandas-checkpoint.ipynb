Different Ways of creating DataFrame
- csv
- excel
- dictionary
- use tuple list

```python
# import pandas as pd
# df=pd.read_csv('name')
# df=pd.read_excel('name')
```

```python
weather_data={
    "day":['1/22/2016','1/25/2016','12/22/2016'],
    "temperature":[34,53,11],
    "windspeed":[33,54,33],
    "Events":['Rain','Sunny','Snow']
}
df=pd.DataFrame(weather_data)
print(df)
```

```python
# List of tuples
data = [
    ("Alice", 24, "USA"),
    ("Bob", 30, "UK"),
    ("Charlie", 22, "India")
]

# Create DataFrame
df = pd.DataFrame(data, columns=["Name", "Age", "Country"]) # yo case ma we must assign column name

print(df)
```

```python
# List of dictionaries
data = [
    {"Name": "Alice", "Age": 24, "Country": "USA"},
    {"Name": "Bob", "Age": 30, "Country": "UK"},
    {"Name": "Charlie", "Age": 22, "Country": "India"}
]

# Create DataFrame
df = pd.DataFrame(data)

print(df)
```

## Read/Write Opeartion in CSV/Excel

```python
df=pd.read_csv('nyc_weather.csv',header=0)
df
## Use 2nd row as header (skipping the first info row)
# df = pd.read_csv('data.csv', header=1)
```

```python
df=pd.read_csv('nyc_weather.csv',nrows=2)
df
```

```python
# Read the CSV file, treating "NaN" and "atulya" as NaN values
df = pd.read_csv('nyc_weather.csv', na_values=["NaN", "atulya"])

# Replace all NaN values with your name
df_filled = df.fillna("Atulya")
df_filled
```

```python
# Read the CSV file, treating "NaN" and "atulya" as NaN values
df = pd.read_csv('nyc_weather.csv', na_values={
     "DewPoint":["NaN", "atulya",-3]    #we can even use dictionary to specify particular column example if sale is in negative which it can't be we can change it to NAN doing like this
    
})

# Replace all NaN values with your name
df_filled = df.fillna("Atulya")
df_filled
```
