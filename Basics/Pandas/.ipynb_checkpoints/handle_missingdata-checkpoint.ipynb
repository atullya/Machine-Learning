```python
df.set_index('day',inplace=True)
df
```

```python
#reaplacing NaN value
#1
new_df=df.fillna(0)
new_df
```

```python
#sometimes 0 is not suitable example in event it make no sense so we do 
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
#if this is also not accepted we can estimate by calcuationg mean, median bvlablalb
#we can also do ffill
new_df=df.fillna(method="ffill")  # yesley chai aagadi ko value j xa teslai copy garey rakdinxa example temperature at first is 32 so second ma NaN ko place ma ni 32 nai rakhxa
# new_df=df.fillna(method="bfill") this oppisite of ffill bhanko backward value fill garxa
new_df
```

```python
new_df=df.fillna(method="bfill",axis="columns") # it is copying value from it's same row i.e horizontally
new_df
```

```python
new_df=df.fillna(method='ffill',limit=1)
new_df
```

```python
#we can also do interpolate (Linear Interpolation) This is one of the better guess
new_df = df.interpolate()

# Now replace NaNs in the 'event' column with 0 (string or number depending on data type)
new_df['event'] = df['event'].fillna(0)

# Print final DataFrame
new_df
```

```python
new_df = df.interpolate(method='time')
new_df #this is more reliable i.e it calculate with time
```

```python
#drop all rows with na values
new_df=df.dropna() # this delete all the row with at least one Na values
neww_df=df.dropna(how='all') # this delete only the row with all Na values
print(neww_df)
new_df
```

```python
n_df = new_df.drop('event',axis=1)
n_df
# #✅ axis=1 means:
# "I want to drop a column."

# ❌ axis=0 would mean:
# "I want to drop a row."
```

```python
data = {
    'name': ['Atulya', 'Rina', 'John', 'Meena', 'David'],
    'score': ['exception', 'good', 'bad', 'exception', 'good']
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
#here score should be in number so we replace it
new_df=df.replace(['exception', 'good', 'bad', 'exception'],[5,3,1,2])
new_df
```

```python
#extra example
data = {
    'name': ['Atulya', 'Rina', 'John', 'Meena', 'David'],
    'score': ['exception', 'good', 'bad', 'exception', 'good']
}

df = pd.DataFrame(data)

# Split
name_col = df['name']
score_col = df['score']

# Operation: Convert scores to uppercase
score_col = score_col.str.upper()

# Merge back
new_df = pd.concat([name_col, score_col], axis=1)

print(new_df)
```
