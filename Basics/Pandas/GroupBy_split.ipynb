```python
import pandas as pd
```

```python
df=pd.read_csv('weather_by_cities.csv')
df
```

```python
#here we do group by cities to do some opeartion
g=df.groupby('city')
g
#interanlly it looks like this
```

```python
for city, city_df in g:
    print(city)
    print(city_df)
```

```python
g.get_group('paris') #get specific group (it is like [Select * from table GROUP BY city])
```

```python
# Create first DataFrame (e.g., student names and their IDs)
df1 = pd.DataFrame({
    'student_id': [1, 2, 3],
    'name': ['Atulya', 'Rina', 'John']
})

# Create second DataFrame (e.g., student IDs and their scores)
df2 = pd.DataFrame({
    'student_id': [1, 2, 3],
    'score': ['good', 'exception', 'bad']
})

# Merge the two DataFrames on 'student_id'
# This will join rows where the 'student_id' matches in both DataFrames
# merged_df = pd.merge(df1, df2)
merged_df = pd.merge(df1, df2, on='student_id')
# Print the result
print(merged_df)
```
