```python
from pymongo import MongoClient
import pandas as pd

# Step 1: Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client['Blog']
collection = db['blogs']

# Step 2: Find documents where title == "Demo"
students1 = list(collection.find({ "title": "Demo" }))

# Step 3: Convert to DataFrame
df_students1 = pd.DataFrame(students1)

# Step 4: Update title to "Demo 1"
df_students1["title"] = "Demo 1"

# Step 5: Remove '_id' to avoid duplicate key error during insert
df_students1.drop(columns=['_id'], inplace=True)

# Step 6: Convert back to dictionary and insert into MongoDB
new_docs = df_students1.to_dict(orient='records')
collection.insert_many(new_docs)

print("Documents updated and inserted successfully.")
```
