```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df
```

```python
# Calculate average rating per book
avg_rating_df = ratings_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()

# Rename the column for clarity
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)

# Show the DataFrame
avg_rating_df
```

```python
#merging two df i.e avg_rating and num_rating
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_ratings']>=230]
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50) # this means top 50 books
popular_df
```

```python
popular_df=popular_df.merge(books,on='Book-Title')
popular_df
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')
popular_df
```

## Collaborative Filtering Based Recommender System

```python
ratings_with_name
```

```python
#we first do filter i.e  we keep the user rating who have done rating for 250 times


```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
x
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index # This is called bookean idexing
padhe_likhe_users
#here in output length=811 means the number of person that have done ratings more than 200 times
```

```python

filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_rating 
```

```python
#now we filter on the basis of book i.e we use only those book which has more than 50 ratings
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt #this pt table mean the books that the user has rate for more than 250 books
```

```python
pt.fillna(0,inplace=True) 
```

```python
#here we have 706 rows Ã— 810 columns now for each books we imagine it has a one point in 810 dimenisional coordinate system
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt) # computes the cosine similarity between all rows of the matrix pt.
print(similarity_scores) #we have calcuated 706  euclidean distance books with 706 books 
print(similarity_scores.shape)
```

```python
#now we create a function that helps to recommend the 5 books for us based on what we search for
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```
