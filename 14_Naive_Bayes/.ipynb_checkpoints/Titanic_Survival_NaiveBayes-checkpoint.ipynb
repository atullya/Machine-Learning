```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
#here Sex column is text we need to convert it into number so we use dummies or ONe hot encoding
dummies=pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
inputs
```

```python
inputs.drop('Sex',axis='columns',inplace=True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()] #to find if any columns as na (missing) value
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
#Now Create a Naive Bayers Model
# There are many of Naive model but we use Gaussian we use Gaussian when our data distribution is normal
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
