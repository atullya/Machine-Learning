```python
stop_words=set(stopwords.words("english"))
stop_words
```

```python
lemmatizer=WordNetLemmatizer()
```

#Main Text Preprocessing Pipeline Function

```python
def preprocess(raw_text):
    # Step 1: lowercase
    lowercase_text = raw_text.lower()

    # Step 2: remove punctuation
    cleaned_text = re.sub(r"[^\w\s]", "", lowercase_text)
    
    # Step 3: tokenize
    tokens = word_tokenize(cleaned_text)
    
    # Step 4: remove stopwords
    filtered_token = [word for word in tokens if word not in stop_words]
    
    # Step 5: lemmatization
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_token]
    
    # return " ".join(lemmatized_tokens)
    clean_text = " ".join(lemmatized_tokens)
    return clean_text
```

```python
preprocess("Distractors are the incorrect answers in a multiple choice question. Choosing your distractors poorly can make your multiple choice questions much easier or harder than you may have intended. Consider the following two questions.")
```
