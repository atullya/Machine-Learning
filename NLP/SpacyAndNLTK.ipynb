```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
doc=nlp("Dr Atullya is a very good person. He is a intelligent person")
doc
```

```python
for sentence in doc.sents:
    print(sentence)
```

```python
text = "Dr Atullya is a very good person. He is a intelligent person"

# Split at period
sentences = text.split(". ")

print(sentences)
print(sentences[0].lower())
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
sent_tokenize("Dr Atullya is a very good person. He is a intelligent person")
```

```python
from nltk.tokenize import word_tokenize
word_tokenize("Dr Atullya is a very good person. He is a intelligent person")
```
