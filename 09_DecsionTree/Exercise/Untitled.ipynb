```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
df.columns
```

```python
inputs=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived'],axis='columns')
inputs
```

```python
targets=df['Survived']
targets
```

```python
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
le_class=LabelEncoder() #4 ota coluns xa tesialey 4 ota object LabelEncoder() ley  convert a specific column into numbers.
le_sex=LabelEncoder()
le_age=LabelEncoder()
le_fare=LabelEncoder()
```

```python
print(inputs['Age'].isnull().sum())
#If it's greater than 0, that means you have missing data.
```

```python
# inputs['encoded_class']=le_class.fit_transform(inputs['Pclass']) # yo garnu parney becauase PClass is already in number
inputs['encoded_sex']=le_sex.fit_transform(inputs['Sex'])
# inputs['encoded_age']=le_age.fit_transform(inputs['Age'])
inputs['encoded_age'] = inputs['Age'].astype(int) 

inputs['encoded_fare']=le_fare.fit_transform(inputs['Fare'])
inputs
```

```python
inputs_n=inputs.drop(['Pclass','Sex','Age','Fare'],axis='columns')
inputs_n
```

```python
model=tree.DecisionTreeClassifier()
model.fit(inputs_n,targets)
```

```python
model.score(inputs_n,targets)
```
