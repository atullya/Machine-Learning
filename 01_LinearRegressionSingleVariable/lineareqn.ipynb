```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 
```

```python
df=pd.read_csv('homeprices.csv') #data frame
df
```

```python
%matplotlib inline
plt.xlabel('area(sq ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
# In machine learning:

# X (Features/Input) = Independent Variable(s) ✅
# y (Target/Output) = Dependent Variable ✅
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price) # X is 2D, y is 1D
## fitting the data means you are training the model using the available data points

## if it runs then our linear regression model is ready to predict the model
```

linear_model.LinearRegression() initializes a Linear Regression model.
-  reg is now an object of the LinearRegression class.
- This object can be used to:
-  Train (Fit) the Model using reg.fit(X, y).
-  Make Predictions using reg.predict(X).
-  Access Model Coefficients (reg.coef_, reg.intercept_).

```python
reg.predict([[3300]])  # Pass as a 2D array

## area aanusar prediction chai kasari garyo bhanne hamro y=mx+b ma value haldai gareko see below
```

```python
#to check how our linear equation looks like
%matplotlib inline
plt.xlabel('area(sq ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue') # y=mx+b

'''
df.area:

This is your x-axis: the independent variable (area in square feet).

reg.predict(df[['area']]):

This uses your trained regression model (reg) to predict prices (y-values) for each corresponding area.

In math terms, it's doing:

𝑦
=
𝑚
⋅
𝑥
+
𝑏
y=m⋅x+b
where:

𝑚
m is the slope (coefficient),

𝑏
b is the intercept.

plt.plot(...):

Plots a line connecting all predicted price points over the x-values (areas).

color='blue':

Sets the line color to blue for visibility.
'''
```
