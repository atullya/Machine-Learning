```python
from sklearn.datasets import load_iris
import pandas as pd
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.data
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
#as per qwn we only consider sepal length and sepal width
```

```python
df=df.drop(['petal length (cm)','petal width (cm)'],axis='columns')
df
```

```python
#now we implementt k-means i.e clustering
from sklearn.cluster import KMeans
```

```python
#renaming for simplicity
```

```python
df.rename(columns={'sepal length (cm)': 'slength', 'sepal width (cm)': 'swidth'})
df
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
plt.scatter(df.slength,df.swidth)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
#this is before we implement k-means
```

```python
km=KMeans(n_clusters=3) 
km #k=3
```

```python
y_predicted=km.fit_predict(df[['slength','swidth']]) #fit and predict both
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
plt.scatter(df1.slength,df1.swidth,color='green')
plt.scatter(df2.slength,df2.swidth,color='blue')

plt.scatter(df3.slength,df3.swidth,color='red')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[['slength']]) 
df['slength'] = scaler.transform(df[['slength']])#it scales the  features

#for age also we scale
scaler.fit(df[['swidth']])            # 2D DataFrame
df['swidth'] = scaler.transform(df[['swidth']])  # result is also 2D, assign back

df
```

```python
#After scaling
km=KMeans(n_clusters=3) 
y_predicted=km.fit_predict(df[['slength','swidth']]) #fit and predict both
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
plt.scatter(df1.slength,df1.swidth,color='green')
plt.scatter(df2.slength,df2.swidth,color='blue')

plt.scatter(df3.slength,df3.swidth,color='red')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.show()
```

```python
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
plt.scatter(df1.slength,df1.swidth,color='green')
plt.scatter(df2.slength,df2.swidth,color='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='+', label='centroid')
plt.scatter(df3.slength,df3.swidth,color='red')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.show()
```

```python
# now we find Elbow Technique i.e like in theory we go throught k=1 to 10 than we find SSE and plot and find that Elbow
k_rng=range(1,10)
sse=[] # sum of square error
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['slength','swidth']])
    sse.append(km.inertia_)  #it is used to give sum of square error
```

```python
sse
```

```python
plt.plot(k_rng, sse, marker='o')
plt.xlabel('K')
plt.ylabel('Sum of Squared Error (SSE)')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```
