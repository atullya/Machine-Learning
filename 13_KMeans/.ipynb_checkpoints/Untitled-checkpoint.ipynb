## IMP NOTE K-means is unsupervised It doesnt need a target label (like spam or not spam, or movie liked or not liked). It just finds patterns or groups (clusters) in the feature space.


 (if k=3 then 3 start we plot)
 (anything in the right is green cluster and on left if red cluster)

Now we try to improve it
The way we try to improve is by adjust the centroid for 2 clusters



 (final cluster)

-> Using Elbow Method i.e we start with k=2 and we try to compute SSE(Sum of Square Error)

we do this same process for k=3,4,5......
Then we draw plot like 

we see that if we incrase the number of k than the error is decreasing
 (as per this chart good will be k=4)

```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
plt.scatter(df.Age,df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
km=KMeans(n_clusters=3) 
km #k=3
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']]) #fit and predict both
y_predicted
```

```python
#Clustring result
df['cluster']=y_predicted
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='blue')

plt.scatter(df3.Age,df3['Income($)'],color='red')
plt.xlabel('Age')
plt.ylabel('Income')

plt.show()
#here in this scatter the problem is  red cluster is not grouped correctly it occur because our scaling is not right
#i.e  hamro y-scale chai 40k dekhi 160k samma gako xa i.e no scaled properly so we use MinMaxScaler()
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])#it scales the income features

#for age also we scale
scaler.fit(df[['Age']])            # 2D DataFrame
df['Age'] = scaler.transform(df[['Age']])  # result is also 2D, assign back

df
```

```python
#again we use k-means algorithm
km=KMeans(n_clusters=3) 
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='blue')

plt.scatter(df3.Age,df3['Income($)'],color='red')
plt.xlabel('Age')
plt.ylabel('Income')

plt.show()
```

```python
km.cluster_centers_ #this is centroid or cetner like that start in example picture
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='blue')

plt.scatter(df3.Age,df3['Income($)'],color='red')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='+', label='centroid') #here x=goes through all the rows i.e 3 row and column of first index and y=all rows column index=1
plt.xlabel('Age')
plt.ylabel('Income')

plt.show()
```

```python
# now we find Elbow Technique i.e like in theory we go throught k=1 to 10 than we find SSE and plot and find that Elbow
k_rng=range(1,10)
sse=[] # sum of square error
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)  #it is used to give sum of square error
```

```python
sse
```

```python
plt.plot(k_rng, sse, marker='o')
plt.xlabel('K')
plt.ylabel('Sum of Squared Error (SSE)')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```
