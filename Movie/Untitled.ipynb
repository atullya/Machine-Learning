```python
def convert_for_cast(text):
    L = []
    counter=0
    for i in ast.literal_eval(text):
        if counter !=3:
            L.append(i['name']) 
            counter+=1
        else:
            break
    return L 
```

```python
 def fetch_director(text):
    L = []
  
    for i in ast.literal_eval(text):
        if i['job'] =='Director':
            L.append(i['name']) 
            break
    return L 
```

```python
# Sam Worthington->SamWorthington (converting to like this because if they are seperate then they may create  two tag example Sam Worthington, it create #Sam and #worthington similary for Sam Watson #Sam Watson
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies
```

```python
# Ensure 'overview' is also converted to list of words
movies['overview'] = movies['overview'].apply(lambda x: x.split())

# Optional: Convert other columns to list of strings (if not already)
# assuming they are lists of dicts with 'name' or strings

# Now create 'tags' by concatenating lists
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

```

```python
new_df=movies[['movie_id','title','tags']]
new_df
```

```python
#list to string for tags
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x)) #It converts a list of words into a single string by joining them with a space.

```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df.head(2)
```

While doing vectorizatin we don't consider stop words eg (to ,from, are,a , in ) i.e the word that is used in sentence formations


```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
text = ["There were atullya is atullya"]
vectors = cv.fit_transform(text)

print(cv.get_feature_names_out())
print(vectors.toarray())
```
