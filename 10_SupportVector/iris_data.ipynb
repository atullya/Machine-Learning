```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris) #dir is  used to obtain a list of valid attributes
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
iris.target
```

```python
df['target']=iris.target
df
```

```python
iris.target_names
```

```python
df[df.target==1].head() #to see how many or which which target value is 1
# df[df.target==0].head()
# df[df.target==2].head()
```

```python
# Add a new column 'flower_name' to the dataframe.
# It takes the 'target' column (which contains numbers like 0, 1, 2),
# and converts them into actual flower names using iris.target_names.
# For example, 0 becomes 'setosa', 1 becomes 'versicolor', etc.
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
df
```
